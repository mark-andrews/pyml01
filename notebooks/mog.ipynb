# Probabilistic mixture models

## Mixture of Gaussians (mog)


```python
import pandas as pd
from sklearn.mixture import GaussianMixture, BayesianGaussianMixture
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
import numpy as np
```

```python
faithful = pd.read_csv("https://raw.githubusercontent.com/mark-andrews/pyml01/master/data/faithful.csv")
```

```python
faithful
```

```python
plt.figure(figsize = (12, 12))
plt.scatter(faithful.values[:,0], faithful.values[:,1]);
plt.xlabel('eruptions')
plt.ylabel('waiting')
```

```python
X_train, X_test = train_test_split(faithful, test_size = 0.15, random_state = 42424)
```

```python
X_train.shape, X_test.shape
```

```python
mog = GaussianMixture(n_components=2)
```

```python
mog.fit(X_train)
```

```python
plt.figure(figsize = (12, 12))
plt.scatter(X_train.values[:,0], X_train.values[:,1], c = mog.predict(X_train));
```

```python
mog.predict(X_test)
```

```python
plt.figure(figsize = (12, 12))
plt.scatter(X_test.values[:,0], X_test.values[:,1], c = mog.predict(X_test));
```

```python
mog.predict_proba(X_train).round(3)
```

```python
p = mog.predict_proba(X_train)[:,0]

plt.figure(figsize = (12, 12))
plt.scatter(X_train.values[:,0], X_train.values[:,1], c = np.log(p));
```

```python
p = mog.predict_proba(X_train)[:,0]

plt.figure(figsize = (12, 12))
plt.scatter(X_train.values[:,0], X_train.values[:,1], c = p);
```

```python
p.round(3)
```
