## clustering moons

```python
X, y = datasets.make_moons(n_samples=1000, noise = 0.1, random_state=42)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 10101)
```

```python
plt.scatter(X[:, 0], X[:, 1], c = y)
```

```python
plt.scatter(X_train[:, 0], X_train[:, 1])
```

```python
kmeans = KMeans(n_clusters=2)
```

```python
kmeans.fit(X_train)
```

```python
y_pred = kmeans.predict(X_test)
```

```python
plt.scatter(X_test[:,0], X_test[:, 1], c = y_pred)
```

```python
plt.scatter(X_test[:,0], X_test[:, 1])
```

```python
kmeans.inertia_
```

```python
silhouette_score(X_test, y_pred)
```

```python
inertia = []
silhouette = []

for k in range(2, 11):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X_train)
    y_pred = kmeans.predict(X_test)
    print('K=%d, score=%2.2f, %2.2f' % (k, kmeans.inertia_, silhouette_score(X_test, y_pred)))
    inertia.append(kmeans.inertia_)
    silhouette.append(silhouette_score(X_test, y_pred))
```

## Clustering digits

```python
mnist = datasets.fetch_openml('mnist_784', version = 1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(mnist['data'],
                                                    mnist['target'],
                                                    test_size = 0.2,
                                                    random_state = 10101)
```

```python
X_train.shape
```

```python
kmeans = KMeans(n_clusters=10)
```

```python
kmeans.fit(X_train)
```

```python
y_pred = kmeans.predict(X_test)
```

```python
X_cluster_5 = X_test[y_pred == 5]
```

```python
fig = plt.figure(figsize = (15, 15))

for i in range(100):
    fig.add_subplot(10, 10, i+1, xticks = [], yticks = [])
    plt.imshow(X_cluster_5[i].reshape(28, 28), cmap = plt.cm.binary, interpolation='nearest')
```

```python
%rep 66
```

```python
def plot_digits_in_cluster(k):
    
    X_cluster = X_test[y_pred == k]

    fig = plt.figure(figsize = (15, 15))

    for i in range(100):
        fig.add_subplot(10, 10, i+1, xticks = [], yticks = [])
        plt.imshow(X_cluster[i].reshape(28, 28), cmap = plt.cm.binary, interpolation='nearest')
```

```python
plot_digits_in_cluster(k = 0)
```
